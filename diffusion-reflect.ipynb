```python
'''
http://www.doitpoms.ac.uk/tlplib/diffusion/fick2.php
'''

def C(x,t, C0=0, Cs=1, D=1, sourceloc=0):
    return Cs - (Cs-C0)*erf((x-sourceloc)/(2*(D*t)**(1/2)))
```

```python
x=np.arange(0,1,.1)
t=np.arange(1,100,1)
```

```python
conc=[]
for time in t:
    conc.append([C(xcoord,time,sourceloc=.1) for xcoord in x])
```

```python
for ctime in conc:
    plt.plot(x,ctime);
plt.xlabel('x');
plt.ylabel('concentration');
```
